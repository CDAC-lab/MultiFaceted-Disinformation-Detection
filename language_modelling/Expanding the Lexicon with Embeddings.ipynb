```python
# Importing Libraries


from sentence_transformers import SentenceTransformer ,  util
import torch
import pandas as pd
import numpy as np
import string
import re
from tqdm.auto import tqdm
tqdm.pandas()

import pickle
```

```python
# Import the sentence transformer model
# model = SentenceTransformer('ghanashyamvtatti/roberta-fake-news', device = 'cuda')
model = SentenceTransformer('sentence-transformers/paraphrase-xlm-r-multilingual-v1', device = 'cuda')
```

```python
def generate_embeddings(words):
  
  embeddings = model.encode(words)
  embeddings=[torch.from_numpy(item) for item in embeddings]
  
  return embeddings
```

```python
with open(r"/content/drive/Shareddrives/[FYP] Fake News Detection/Kogul_Language_Modelling/Lexicon Generation/wordlist.pickle", "rb") as input_file:
  wordlist = pickle.load(input_file)
```

```python
businessf, businessr = pd.DataFrame(), pd.DataFrame()
educationf, educationr = pd.DataFrame(), pd.DataFrame()
sportf, sportr = pd.DataFrame(), pd.DataFrame()
entertainmentf, entertainmentr = pd.DataFrame(), pd.DataFrame()
politicalf, politicalr = pd.DataFrame(), pd.DataFrame()
technologyf, technologyr = pd.DataFrame(), pd.DataFrame()
```

```python
business_fake, business_real = wordlist['business']['fake'], wordlist['business']['true']
education_fake, education_real = wordlist['education']['fake'], wordlist['education']['true'] 
sport_fake, sport_real = wordlist['sport']['fake'], wordlist['sport']['true'] 
entertainment_fake, entertainment_real = wordlist['entertainment']['fake'], wordlist['entertainment']['true'] 
political_fake, political_real = wordlist['political']['fake'], wordlist['political']['true'] 
technology_fake, technology_real = wordlist['technology']['fake'], wordlist['technology']['true']  
```

```python
businessf_embeddings, businessr_embeddings = generate_embeddings(business_fake), generate_embeddings(business_real)
educationf_embeddings, educationr_embeddings = generate_embeddings(education_fake), generate_embeddings(education_real)
sportf_embeddings, sportr_embeddings = generate_embeddings(sport_fake), generate_embeddings(sport_real)
entertainmentf_embeddings, entertainmentr_embeddings = generate_embeddings(entertainment_fake), generate_embeddings(entertainment_real)
politicalf_embeddings, politicalr_embeddings = generate_embeddings(political_fake), generate_embeddings(political_real)
technologyf_embeddings, technologyr_embeddings = generate_embeddings(technology_fake), generate_embeddings(technology_real)
```

```python
businessf['word'], businessf['embedding'], businessf['label'] = business_fake, businessf_embeddings, 1
businessr['word'], businessr['embedding'], businessr['label'] = business_real, businessr_embeddings, 0

educationf['word'], educationf['embedding'], educationf['label'] = education_fake, educationf_embeddings, 1
educationr['word'], educationr['embedding'], educationr['label'] = education_real, educationr_embeddings, 0

sportf['word'], sportf['embedding'], sportf['label'] = sport_fake, sportf_embeddings, 1
sportr['word'], sportr['embedding'], sportr['label'] = sport_real, sportr_embeddings, 0

entertainmentf['word'], entertainmentf['embedding'], entertainmentf['label'] = entertainment_fake, entertainmentf_embeddings, 1
entertainmentr['word'], entertainmentr['embedding'], entertainmentr['label'] = entertainment_real, entertainmentr_embeddings, 0

politicalf['word'], politicalf['embedding'], politicalf['label'] = political_fake, politicalf_embeddings, 1
politicalr['word'], politicalr['embedding'], politicalr['label'] = political_real, politicalr_embeddings, 0

technologyf['word'], technologyf['embedding'], technologyf['label'] = technology_fake, technologyf_embeddings, 1
technologyr['word'], technologyr['embedding'], technologyr['label'] = technology_real, technologyr_embeddings, 0
```

```python
# Stack the DataFrames on top of each other
business = businessf.append(businessr,ignore_index = True)
education = educationf.append(educationr,ignore_index = True)
sport = sportf.append(sportr,ignore_index = True)
entertainment = entertainmentf.append(entertainmentr,ignore_index = True)
political = politicalf.append(politicalr,ignore_index = True)
technology = technologyf.append(technologyr,ignore_index = True)
```
