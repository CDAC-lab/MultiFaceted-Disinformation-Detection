```python
import pandas as pd
from annoy import AnnoyIndex
```

```python
dfbusiness = pd.read_pickle("/content/drive/Shareddrives/[FYP] Fake News Detection/Kogul_Language_Modelling/Lexicon Generation/Word Embeddings/business.pickle")
dfeducation = pd.read_pickle("/content/drive/Shareddrives/[FYP] Fake News Detection/Kogul_Language_Modelling/Lexicon Generation/Word Embeddings/education.pickle")
dfentertainment = pd.read_pickle("/content/drive/Shareddrives/[FYP] Fake News Detection/Kogul_Language_Modelling/Lexicon Generation/Word Embeddings/entertainment.pickle")
dfpolitical = pd.read_pickle("/content/drive/Shareddrives/[FYP] Fake News Detection/Kogul_Language_Modelling/Lexicon Generation/Word Embeddings/political.pickle")
dfsport = pd.read_pickle("/content/drive/Shareddrives/[FYP] Fake News Detection/Kogul_Language_Modelling/Lexicon Generation/Word Embeddings/sport.pickle")
dftechnology = pd.read_pickle("/content/drive/Shareddrives/[FYP] Fake News Detection/Kogul_Language_Modelling/Lexicon Generation/Word Embeddings/technology.pickle")
```

```python
business_embeddings = dfbusiness['embedding'].tolist()
education_embeddings = dfeducation['embedding'].tolist()
entertainment_embeddings = dfentertainment['embedding'].tolist()
political_embeddings = dfpolitical['embedding'].tolist()
sport_embeddings = dfsport['embedding'].tolist()
technology_embeddings = dftechnology['embedding'].tolist()
```

```python
f = 768
```

```python
business = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for i in range(len(business_embeddings)):
    business.add_item(i, business_embeddings[i])

education = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for i in range(len(education_embeddings)):
    education.add_item(i, education_embeddings[i])

entertainment = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for i in range(len(entertainment_embeddings)):
    entertainment.add_item(i, entertainment_embeddings[i])

political = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for i in range(len(political_embeddings)):
    political.add_item(i, political_embeddings[i])

sport = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for i in range(len(sport_embeddings)):
    sport.add_item(i, sport_embeddings[i])

technology = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for i in range(len(technology_embeddings)):
    technology.add_item(i, technology_embeddings[i])
```

```python
business.build(1000, n_jobs=-1) # 1000 trees
education.build(1000, n_jobs=-1)
entertainment.build(1000, n_jobs=-1)
political.build(1000, n_jobs=-1)
sport.build(1000, n_jobs=-1)
technology.build(1000, n_jobs=-1)
```
